```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
school_data_complete["budget"].head()
```

```python
schools = len(school_data_complete["school_name"].unique())

students = school_data_complete["Student ID"].count()
# Getting the Budget for each school
grouped_hs_df = school_data_complete.groupby(["school_name"])
budget_school = grouped_hs_df["budget"].sum()
total_budget = budget_school.sum()
#total_budget = school_data_complete["budget"].sum()

average_math_score = school_data_complete["math_score"].mean()

average_reading_score = school_data_complete["reading_score"].mean()

passing_math = (school_data_complete["math_score"] >= 70).sum()
#passing_math_count = passing_math.count()
percent_passing_math = (passing_math / students) * 100

passing_reading = (school_data_complete["reading_score"] >= 70).sum()
#passing_math_count = passing_math.count()
percent_passing_reading = (passing_reading / students) * 100

passed_both = ((school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)).sum()
percent_passing_both = (passed_both / students) * 100

district_summary_df = pd.DataFrame({"Total Schools": [schools],
                              "Total Students": students,
                              "Total Budget": total_budget,
                              "Average Math Score": average_math_score,
                              "Average Reading Score": average_reading_score,
                                   "Passing Math": percent_passing_math,
                                   "Passing Reading": percent_passing_reading,
                                   "Overall Passing": percent_passing_both})
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:

```python
#students_per_hs = school_data_complete.groupby('school_name').count()
#grouped_hs_df = school_data_complete.groupby(["school_name"],as_index=False)
#print(grouped_hs_df)
#grouped_hs_df.count().head(10)
school_name = school_data_complete["school_name"].unique()

students_per_hs = school_data_complete.groupby("school_name")['Student ID'].count()
#print (students_per_hs)
school_type = school_data_complete.groupby('school_name')['type'].unique()
school_budget = school_data_complete.groupby("school_name")["budget"].unique()
student_budget = school_budget/students_per_hs

school_avg_math = school_data_complete.groupby('school_name')['math_score'].mean()
school_avg_read = school_data_complete.groupby('school_name')['reading_score'].mean()

reading_pass_df = school_data_complete[school_data_complete["reading_score"] > 70]
school_pass_read = reading_pass_df.groupby('school_name')['Student ID'].count()
math_pass_df = school_data_complete[school_data_complete["math_score"] > 70]
school_pass_math = math_pass_df.groupby('school_name')['Student ID'].count()

percent_school_pass_read = (school_pass_read / students_per_hs) * 100
percent_school_pass_math = (school_pass_math / students_per_hs) * 100

#pass_both_df = school_data_complete[school_data_complete["math_score"] > 70] & [school_data_complete["reading_score"] > 70]
pass_both_df = school_data_complete.loc[(school_data_complete['math_score'] > 70) & (school_data_complete['reading_score'] > 70)]
school_pass_both = pass_both_df.groupby('school_name')['Student ID'].count()
#passed_school_both = ((school_data_complete.groupby('school_name')["reading_score"] >= 70) & (school_data_complete.groupby('school_name')["math_score"] >= 70)).sum()
percent_school_both = (school_pass_both / students_per_hs) * 100


school_summary_df = pd.DataFrame({"School Type": school_type,
                                 "Total Students": students_per_hs,
                                 "Total School Budget": school_budget,
                                 "Student Budget": student_budget,
                                 "Average Math Score": school_avg_math,
                                 "Average Reading Score": school_avg_read,
                                 "% Passing Math": percent_school_pass_math,
                                 "% Passing Reading": percent_school_pass_read,
                                 "% Overall Passing": percent_school_both})
#school_summary_df.reset_index()
school_summary_df["Total School Budget"] = school_summary_df.loc[:, "Total School Budget"].astype(float)
school_summary_df.head()
```
